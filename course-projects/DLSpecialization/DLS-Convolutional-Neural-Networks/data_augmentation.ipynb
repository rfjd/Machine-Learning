# Data augmentation by Keras
There are many more data augmentation methods. Please visit the documentation [website](https://www.tensorflow.org/tutorials/images/data_augmentation) for more info.

```python
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import tensorflow as tf
tfl = tf.keras.layers
```

```python
img = image.load_img('./data/hamster.jpg')
img = image.img_to_array(img)/255
def PLOT(img,title):
    plt.figure(figsize=(4,4))
    plt.imshow(img)
    plt.title(title)
    plt.show()
    
PLOT(img,'original')
```

## Resize

```python
resize_layer = tf.keras.Sequential([tfl.Resizing(512,512)])
img_resized = resize_layer(img)
PLOT(img_resized,'resized')
```

## Random Flip

```python
flip_layer = tf.keras.Sequential([tfl.RandomFlip(mode="horizontal_and_vertical")])
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(12, 3))
fig.subplots_adjust(wspace=0.2, hspace=0.1)
for a in ax.flat:
    img_flipped = flip_layer(img)
    a.imshow(img_flipped)
    a.set_axis_off()
    
plt.show()
```

## Random Rotation

```python
rotate_layer = tf.keras.Sequential([tfl.RandomRotation(factor=0.25)])
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(12, 3))
fig.subplots_adjust(wspace=0.2, hspace=0.1)
for a in ax.flat:
    img_rotated = rotate_layer(img)
    a.imshow(img_rotated)
    a.set_axis_off()
    
plt.show()
```

## Multiple methods

```python
layer = tf.keras.Sequential([flip_layer,rotate_layer])
img_modified = layer(img)
PLOT(img_modified,'modified')
```

## Polynomial regression

```python
%run ../../src.ipynb

x = np.arange(0,10,1)
y = np.power(x,2)+1+np.random.rand(x.size)
xPolynomial = np.c_[x,np.power(x,2),np.power(x,3)]

# feature scaling
mu = np.mean(xPolynomial, axis=0)
sigma = np.std(xPolynomial, axis=0)

xScaled = (xPolynomial-mu)/sigma

m, n = xScaled.shape;
w0 = np.zeros(n)
b0 = 0
alpha = 1e-1
numberIterations = 20000
w,b,J = gradient_descent(xScaled,y,w0,b0,alpha,numberIterations,linear_regression_cost,linear_regression_gradient);
print(f"w = {w}, b = {b}")
yhat = prediction_linear_regression(w,b,xScaled)
plt.scatter(x,y,marker='o',c='r',label='training data')
plt.plot(x,yhat,'-b',label='model')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(frameon=False)
plt.show()
```

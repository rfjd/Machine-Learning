## Logistic regression for university admission

### Load the data and perform a preliminary investigation

```python
%run ../src.ipynb

data = np.loadtxt("../../data/examScores.txt", delimiter=',')
x = data[:,0:-1]
y = data[:,-1]

m, n = x.shape
print(f"m = {m} training examples")
print(f"n = {n} features")

def PLOT(x):
    pos = y == 1
    neg = y != 1

    plt.scatter(x[pos,0],x[pos,1],label='admitted')
    plt.scatter(x[neg,0],x[neg,1],label='rejected')
    plt.xlabel("exam \#1 (\%)")
    plt.ylabel("exam \#2 (\%)")
    plt.legend(frameon=False)
    
PLOT(x)
plt.show()
```

### Use gradient descent to find the best linear fit for the original data

```python
np.random.seed(1)
w0 = 0.01 * (np.random.rand(2) - 0.5)
b0 = -8
alpha = 1e-3 # very small!
numberIterations = 10000
w,b,J = gradient_descent(x,y,w0,b0,alpha,numberIterations,logistic_regression_cost,logistic_regression_gradient);
print(f"w = {w}, b = {b}")

yhat = prediction_logistic_regression(w,b,x)

PLOT(x)
# decision boundary is given by z = 0
plt.plot([0,-b/w[1]],[-b/w[0],0])
plt.xlim([0,100])
plt.ylim([0,100])
plt.show()

# accuracy:
print(f"The training accuracy is {np.mean(yhat == y)*100}%")

```
